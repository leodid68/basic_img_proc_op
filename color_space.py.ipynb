```python
def plt_imshow(name, chan):
    plt.figure()
    plt.imshow(chan, cmap='gray')
    plt.title(f'Channel {name}')
    plt.axis('off')
    plt.show()
```

# Analyse du code de traitement d'image

## Description générale
Ce code charge et traite une image en utilisant OpenCV et Matplotlib pour visualiser les différents canaux de couleur (RGB).

## Détails des étapes

### 1. Importation des bibliothèques
- `matplotlib.pyplot` : Pour l'affichage des images et des graphiques
- `numpy` : Pour les calculs sur les tableaux
- `cv2` : Pour le traitement d'image via OpenCV
- `IPython.display` : Pour l'affichage interactif dans le notebook

### 2. Chargement de l'image
- L'image est chargée depuis le chemin spécifié 
- Affichage de l'image d'origine avec `display(Image())`

### 3. Séparation et visualisation des canaux RGB 
Le code décompose l'image en ses trois canaux de couleur :
- B (Bleu)
- G (Vert) 
- R (Rouge)

Pour chaque canal :
- Création d'une nouvelle figure
- Affichage du canal en niveaux de gris
- Ajout d'un titre identifiant le canal
- Désactivation des axes
- Affichage du résultat

## Utilité
Cette décomposition permet :
- D'analyser la contribution de chaque couleur dans l'image
- De comprendre la composition colorimétrique
- D'identifier les zones dominantes pour chaque canal
- De faciliter certains traitements spécifiques à un canal


```python
for (name, chan) in zip(("B", "G", "R"), cv2.split(image)):
    plt.figure()
    plt.imshow(chan, cmap='gray')
    plt.title(f'Channel {name}')
    plt.axis('off')
    plt.show()
```

# Pourquoi étudier les espaces de couleurs en traitement d'image ?

## L'espace de couleur HSV

HSV (Hue, Saturation, Value) est un espace de couleur alternatif à RGB qui est particulièrement utile en vision par ordinateur :

1. **Teinte (Hue)** :
   - Représente la couleur pure
   - Valeur entre 0° et 360° sur le cercle chromatique
   - Facilite l'identification et le filtrage des couleurs

2. **Saturation (Saturation)** :
   - Mesure la pureté/intensité de la couleur
   - De 0% (gris) à 100% (couleur pure)
   - Utile pour détecter les zones colorées vs neutres

3. **Valeur (Value)** : 
   - Représente la luminosité
   - De 0% (noir) à 100% (lumineux)
   - Permet d'analyser les variations d'éclairage

### Avantages en vision par ordinateur :

- Meilleure robustesse aux changements d'illumination
- Séparation intuitive de l'information de couleur et de luminosité
- Idéal pour la segmentation basée sur la couleur
- Très utilisé en détection d'objets et tracking


```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

plt_imshow("HSV", hsv)
```

```python
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
plt_imshow("LAB", lab)
```

```python
for (name, chan) in zip(("L*", "a*", "b*"), cv2.split(lab)):
    plt_imshow(name, chan)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt_imshow("Gray", gray)
```
