# Pourquoi tudier les espaces de couleurs en traitement d'image ?

## L'espace de couleur HSV

HSV (Hue, Saturation, Value) est un espace de couleur alternatif  RGB qui est particulirement utile en vision par ordinateur :

1. **Teinte (Hue)** :
   - Reprsente la couleur pure
   - Valeur entre 0 et 360 sur le cercle chromatique
   - Facilite l'identification et le filtrage des couleurs

2. **Saturation (Saturation)** :
   - Mesure la puret/intensit de la couleur
   - De 0% (gris)  100% (couleur pure)
   - Utile pour dtecter les zones colores vs neutres

3. **Valeur (Value)** : 
   - Reprsente la luminosit
   - De 0% (noir)  100% (lumineux)
   - Permet d'analyser les variations d'clairage

### Avantages en vision par ordinateur :

- Meilleure robustesse aux changements d'illumination
- Sparation intuitive de l'information de couleur et de luminosit
- Idal pour la segmentation base sur la couleur
- Trs utilis en dtection d'objets et tracking


```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

plt_imshow("HSV", hsv)
```

```python
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
plt_imshow("LAB", lab)
```

```python
for (name, chan) in zip(("L*", "a*", "b*"), cv2.split(lab)):
    plt_imshow(name, chan)
```
