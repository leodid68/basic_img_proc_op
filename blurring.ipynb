```python
image_file = "/Users/leodidier/Documents/GitHub/basic_img_proc_op/images/radeau.jpg"
image_path = "/Users/leodidier/Documents/GitHub/basic_img_proc_op/images"
```

```python
image = cv2.imread(image_file)
```

```python
display(Image(filename=image_file))
```

```python
kernelSizes = [(3, 3), (9,9), (15, 15)]
```

# Explication de la Boucle avec cv2.blur

Cette boucle applique un flou (blur)  l'image avec diffrentes tailles de noyau :

1. La boucle parcourt la liste `kernelSizes` qui contient 3 paires de valeurs: (3,3), (9,9), et (15,15)
2. Pour chaque paire (kX, kY):
   - `cv2.blur()` applique un filtre de flou moyen avec une fentre de taille kX  kY pixels
   - Plus la taille du noyau est grande, plus l'effet de flou est prononc
   - L'image floute est sauvegarde avec un nom indiquant la taille du noyau utilis
   - L'image rsultante est affiche dans le notebook

Le flou moyen fonctionne en remplaant chaque pixel par la moyenne de tous les pixels dans la fentre kX  kY centre sur ce pixel.


```python
for (kX, kY) in kernelSizes:
    blurred = cv2.blur(image, (kX, kY))
    cv2.imwrite(os.path.join(image_path, f"blurred_{kX}_{kY}.jpg"), blurred)
    print(f"Dimensions du noyau: ({kX}, {kY})")
    display(Image(filename=os.path.join(image_path, f"blurred_{kX}_{kY}.jpg")))
    print(f"kX: {kX}, kY: {kY}")
```

# Application du Flou Gaussien (GaussianBlur)

Cette boucle applique un flou gaussien  l'image avec diffrentes tailles de noyau :

1. La boucle parcourt la liste `kernelSizes` contenant 3 tailles de noyau : (3,3), (9,9), et (15,15)
2. Pour chaque paire (kX, kY) :
   - `cv2.GaussianBlur()` applique un filtre de flou gaussien avec une fentre de taille kX  kY pixels
   - Les paramtres sigmaX=0 et sigmaY=0 permettent  OpenCV de calculer automatiquement les carts-types optimaux
   - L'image floute est sauvegarde avec un nom indiquant la taille du noyau utilis
   - L'image rsultante est affiche dans le notebook
   - Les dimensions du noyau utilis sont affiches

Contrairement au flou moyen, le flou gaussien utilise une moyenne pondre o les pixels plus proches du centre ont plus d'importance que ceux en priphrie. Cette pondration suit une distribution gaussienne, ce qui donne un effet de flou plus naturel et plus doux.


```python
for (kX, kY) in kernelSizes:
    gaussblurred = cv2.GaussianBlur(image, (kX, kY), sigmaX=0, sigmaY=0)
    cv2.imwrite(os.path.join(image_path, f"gaussblurred_{kX}_{kY}.jpg"), gaussblurred)
    print(f"Dimensions du noyau: ({kX}, {kY})")
    display(Image(filename=os.path.join(image_path, f"gaussblurred_{kX}_{kY}.jpg")))
    print(f"kX: {kX}, kY: {kY}")
```
